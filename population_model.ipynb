```python
# 必要なライブラリのインポート
import numpy as np
import matplotlib.pyplot as plt

# 初期パラメータ
initial_population = 1000   # 初期人口
growth_rate = 0.02  # 成長率 (per year)
time_steps = 100    # シミュレーション期間 (years)

# 時間と人口の配列を初期化
time = np.arange(time_steps)
population = np.empty(time_steps)

# 初期人口を設定
population[0] = initial_population

# 人口の成長をシミュレーション
for t in range(1, time_steps):
    population[t] = population[t-1] * (1 + growth_rate)

# 結果をプロット
plt.figure(figsize=(10,6))
plt.plot(time, population)
plt.xlabel('Time (years)')
plt.ylabel('Population')
plt.title('Exponential Population Growth')
plt.grid(True)
plt.show()
```
